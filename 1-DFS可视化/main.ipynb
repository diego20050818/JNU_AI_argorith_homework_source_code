```python
#定义起始点和目的地
start = '珠海'
end = '北京'
#stack = [(start,[start])]
#visited = set()
#all_path = []
path = []
stack = []

count_range = 0

def dfs_search_path(point=start, end=end, G=city_graph):
    global count_range
    count_range += 1
    stack.append(point)
    if point == end:
        print(f"寻找到从{start}到{end}的路径:",end="")
        for i in stack[:len(stack)-1]:
            print(i+" -> ",end="")
        print(stack[len(stack)-1],end="")
        
        path.append(list(stack))
        stack.pop()
        print(f"   迭代次数：{count_range}")
        return
    for neighbor in G[point]:
        if neighbor not in stack:
            dfs_search_path(neighbor, end, G)
    stack.pop()
    print(f"正在寻找路径：{stack}",end="\r")
    sys.stdout.flush()
    #time.sleep(0.01)
    
    

try: 
    dfs_search_path()
    count = []
    for i in path:
        count.append(len(i))
    min_index = count.index(min(count))
    #print(f"{start}到{end}最短路径为{path[min_index]},长度为{len(path[min_index])-1}")
    min_path = ""
    for j in path[min_index][:len(path[min_index])-1]:
        min_path += f"{j} -> "
    min_path += end
    print(f"\n找到{len(path)}条路径")
    print(f"{start}到{end}最短路径为{min_path},长度为{len(path[min_index])-1}")
except Exception as e:
    print(f"输入图像不是全连通图或者{e}")



  
```
